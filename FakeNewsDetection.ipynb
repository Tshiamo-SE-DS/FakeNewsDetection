```python
data = merge_data.drop(['title', 'subject', 'date'], axis = 1) #Here we remove columns that are not required going further with the process.
```

```python
data.isnull().sum()
```

```python
data = data.sample(frac = 1) #Here we are randomly shuffling the dataframes.
```

```python
data.head()
```

```python
data.reset_index(inplace = True)
data.drop(['index'], axis = 1, inplace = True)
```

```python
data.columns
```

```python
data.head()
```

```python
def wordopt(text): #Here we create a function to process the text.
    text = text.lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub("\\W", " ", text)
    text = re.sub('https?://\S+|www\.\S+', '', text)
    text = re.sub('<.*?>+', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\n', '', text)
    text = re.sub('\w*\d\w*', '', text)
    return text
```

```python
data['text'] = data['text'].apply(wordopt)
```

```python
x = data['text'] #Here we define the dependent and independent variables.
y = data['class']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.25) #Here we are splitting training and testing data.
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorization = TfidfVectorizer()
xv_train = vectorization.fit_transform(x_train)
xv_test = vectorization.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression #Linear Regression Classification

LR = LogisticRegression()
LR.fit(xv_train, y_train)
```

```python
pred_lr = LR.predict(xv_test)
```

```python
LR.score(xv_test, y_test) #Here we are displaying the Linear Regression accuracy score
```

```python
print(classification_report(y_test, pred_lr)) #Here we are displaying the classification report of the Linear Regression
```

```python
pred_gb = GB.predict(xv_test)
```

```python
GB.score(xv_test, y_test) #Here we are displaying the Gradient Boosting accuracy score
```

```python
print(classification_report(y_test, pred_gb)) #Here we are displaying the classification report of the Gradient Boosting
```

```python
from sklearn.ensemble import RandomForestClassifier #Random Forest Classification
RF = RandomForestClassifier(random_state = 0)
RF.fit(xv_train, y_train)
```

```python
pred_rf = RF.predict(xv_test)
```

```python
RF.score(xv_test, y_test) #Here we are displaying the Random Forest accuracy score
```
